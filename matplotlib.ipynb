```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #Jupyter notebook提供一個方法可以直接在螢幕上秀出劃出的圖表
```

```python
#建立一個矩陣從-10到10(100個等差數值)，x為剛剛建立的矩陣，y為sin()，畫圖的樣式為「.」
ironman = np.linspace(-10,10,100)
fig = plt.figure() #定義圖像窗口
#折線圖
plt.plot(ironman,np.sin(ironman),'.') #定義x,y和圖的樣式
```

```python
#同時畫兩條線
fig = plt.figure()
plt.plot(ironman,np.sin(ironman),'.',ironman,np.cos(ironman),'--')
```

```python
plt.style.use('bmh')
fig=plt.figure()
ax =plt.axes()
plt.plot(ironman,np.sin(ironman),'.')
plt.xlim(-15,15)
plt.ylim(-1.5,1.5)
plt.title("Sine curve y=sin(x)")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
fig = plt.figure()
ironman_hist = np.random.randn(1000)
#直方圖
plt.hist(ironman_hist,bins=20,color='SteelBlue',edgecolor="LightSteelBlue") 
#bins 定義有幾個直方圖 color->直方圖顏色 edgecolor->邊界顏色
```

```python
counts,bin_edges= np.histogram(ironman_hist,bins=20)
print('counts---->',counts)
print('bin_edges--->',bin_edges)
```

```python
fig = plt.figure()
labels =["day1","day2","day3","day4","day5","day6"]
word_size=[2000,3250,4600,2300,3600,5600]
explode =[0,0,.05,0,0,0]
 #0則不突凸出，值越大 則凸出越大

 #圓餅圖
plt.pie(word_size,explode=explode,labels=labels,autopct="%3.1f%%") 
# autopct-->項目百分比的格式
```

```python
x1 = [np.random.randint(0,20) for _ in range(10)]
#返回一個隨機整型數，範圍從低（包括）到高（不包括）
x2 = [np.random.randint(0,20) for _ in range(10)]
y1 = [np.random.randint(30,60) for _ in range(10)]
y2 = [np.random.randint(50,80) for _ in range(10)]
#分散圖
plt.scatter(x1, y1, label='group 1', marker='^', c='r', s=[150 for s in range(len(x1))])
plt.scatter(x2, y2, label='group 2', marker='s', c='c', s=[150 for s in range(len(x2))])

plt.title("Multiple Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show() 
```
